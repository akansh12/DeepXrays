```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from tqdm.notebook import tqdm
from glob import glob
from PIL import Image
```

```python
import torch
from torchvision import utils
```

```python
train_loc = "../input/chestxrgrand/train/"
val_loc = "../input/chestxrgrand/validation"
```

### There are three classes. Covid, Normal and Pneumonia. 

```python
print("Number of folders in train:", os.listdir(train_loc))
print("Number of folders in validation:", os.listdir(val_loc))
```

```python
#train
pneumonia_train = os.listdir(os.path.join(train_loc, os.listdir(train_loc)[0]))
normal_train = os.listdir(os.path.join(train_loc, os.listdir(train_loc)[1]))
covid_train = os.listdir(os.path.join(train_loc, os.listdir(train_loc)[2]))
```

```python
print("Number of images in train Pneumonia: ",len(pneumonia_train))
print("Number of images in train Normal: ",len(normal_train))
print("Number of images in train Covid: ",len(covid_train))
```

```python
# checking if file is repeted
print(len(np.unique(normal_train)))
print(len(np.unique(pneumonia_train)))
print(len(np.unique(covid_train)))
```

```python
plt.figure(figsize=(30,30)) 
for i in range(1,10):
    plt.subplot(1,10,i,)
    plt.imshow(plt.imread((os.path.join(train_loc+'normal', normal_train[i]))), cmap = 'gray')
    plt.title(f"Normal {i}")
plt.figure(figsize=(30,30)) 
for i in range(1,10):
    plt.subplot(1,10,i,)
    plt.imshow(plt.imread((os.path.join(train_loc+'covid', covid_train[i]))), cmap = 'gray')
    plt.title(f"Covid {i}")
plt.figure(figsize=(30,30)) 
for i in range(1,10):
    plt.subplot(1,10,i,)
    plt.imshow(plt.imread((os.path.join(train_loc+'pneumonia', pneumonia_train[i]))), cmap = 'gray')
    plt.title(f"pneumonia {i}")
```
